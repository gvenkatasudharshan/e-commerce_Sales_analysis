```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Importing the data filein python using below path
file_path = r'E:\Python\E-commmerce\e-commerce.csv'
# to read the file based on the extention need use the method like read_csv() ,read_xlxs()
data_set=pd.read_csv(file_path)
# by using info() funtion to get the complete information file, we the get the column names,count of columns, data types of columns
data_set.info()
```

```python
# by using describe() function, we will get the overall count , mean, std, min, max this funtion use the only numeric values in the file
#to preform the funtion 
data_set.describe()
```

```python
# is null() is used to check the null value(empty cells) having in the data, sum() gives the total null values having in each column
data_set.isnull().sum()
```

```python
# head() is used to find the first rows in the data based on the condition
data_set.head(20)
```

```python
# tail()is used to find the last rows based on the condition 
data_set.tail(20)
```

```python
# by using this loop we able to display the complete data in file 
tab_size=20
for i in range(0,len(data_set),tab_size):
    display(data_set[i:i+tab_size])
```

```python
# here converting the order date into date format and separate the month and year format
data_set['Order_date'] = pd.to_datetime(data_set['Order_date'])
data_set['month_year'] = data_set['Order_date'].apply(lambda x : x.strftime('%Y-%m'))
print(data_set['month_year'])
```

```python
# grouping the data set based on month year 
sales_trend = data_set.groupby('month_year').sum('Sales').reset_index()
display(sales_trend)
# creating the graph  based on the  monthly  sales
plt.figure(figsize=(12,6))
plt.title('Montly Sales')
bar1=plt.bar(sales_trend['month_year'],sales_trend['Sales'])
for bars in bar1:
    height = bars.get_height()
    plt.text(bars.get_x() + bars.get_width() / 2, height, f'{height}', ha='center', va='bottom')
plt.show()
```

```python
# finding the top three months sales  and reperesenting in graph
top_three_months=sales_trend.sort_values(by='Sales', ascending=False).head(3)
plt.figure(figsize=(4,5))
plt.title('Top three Months in Sales')
bar2=plt.bar(top_three_months['month_year'],top_three_months['Sales'])
plt.xticks(rotation=90)
for bars in bar2:
    height = bars.get_height()
    plt.text(bars.get_x() + bars.get_width() / 2, height, f'{height}', ha='center', va='bottom')
plt.show()
```

```python
# grouping the data based on the category 
ctg_trend = data_set.groupby('Category', as_index =False).sum('Sales')
display(ctg_trend)
```

```python
# representing category wise sales with bar and line graph both combination
plt.figure(figsize=(20,10))
plt.plot(ctg_trend['Category'],ctg_trend['Sales'],color='#FF5733')
bar=plt.bar(ctg_trend['Category'],ctg_trend['Sales'])
plt.title('Categroy wise Sales')
plt.xticks(rotation=90)
for bars in bar:
    height = bars.get_height()
    plt.text(bars.get_x() + bars.get_width() / 2, height, f'{height}', ha='center', va='bottom')

plt.show()
```

```python
# finding the category wise stock Quantity and reperesenting in graph
plt.figure(figsize=(20,10))
plt.title('Categroy wise Stock Quantity')
plt.scatter(ctg_trend['Category'],ctg_trend['StockQuantity'] ,color='red')
plt.plot(ctg_trend['Category'],ctg_trend['StockQuantity'] ,color='red')
bar=plt.bar(ctg_trend['Category'],ctg_trend['StockQuantity'])
plt.xticks(rotation=90)
for bars in bar:
    height = bars.get_height()
    plt.text(bars.get_x() + bars.get_width() / 2, height, f'{height}', ha='center', va='bottom')
plt.show()
```

```python
# finding the top three categories sales and reperesenting in graph
top_three=ctg_trend.sort_values(by='Sales', ascending=False).head(3)
plt.figure(figsize=(4,5))
plt.title('Top Three Categries in Sales')
bar2=plt.bar(top_three['Category'],top_three['Sales'])
plt.xticks(rotation=90)
for bars in bar2:
    height = bars.get_height()
    plt.text(bars.get_x() + bars.get_width() / 2, height, f'{height}', ha='center', va='bottom')
plt.show()
```

```python
# finding the last three categories sales and reperesenting in graph
last_three=ctg_trend.sort_values(by='Sales', ascending=False).tail(3)
plt.figure(figsize=(4,5))
plt.title('Least Three Categries in Sales')
bar2=plt.bar(last_three['Category'],last_three['Sales'])
plt.xticks(rotation=90)
for bars in bar2:
    height = bars.get_height()
    plt.text(bars.get_x() + bars.get_width() / 2, height, f'{height}', ha='center', va='bottom')
plt.show()
```
